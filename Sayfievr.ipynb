```python
df = pd.read_excel(open ('uniform.xlsx', 'rb'))
df.head()
```

```python
cols = ["Name", "Number", "Store", "Size"]
df[cols].sort_values("Number", ascending=False) #проба
```

```python
df.Number.sum() # общее количество униформы требуемой для резервирования места на складе
```

```python
df.groupby('Store').aggregate({'Number':'sum'}) # заказ по магазинам - общее количество
```

```python
df.groupby('Store').aggregate({'Number':'sum'}).sort_values("Number", ascending=False) 
# сортировка по большему спросу - приоритезация по магазинам по готовности униформы
```

```python
df.groupby('Name').aggregate({'Number':'sum'}).sort_values("Number", ascending=False) 
# доля количества униформы для отдела кадров
group1 = df.groupby('Name').aggregate({'Number':'sum'}).sort_values("Number", ascending=False)
```

```python
df[cols].groupby(['Name','Size']).aggregate({'Number':'sum'}).sort_values("Number", ascending=False) 
#группировка униформы по размеру и количеству без учета магазинов для проведения тендера
```
